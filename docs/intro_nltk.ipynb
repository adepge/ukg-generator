#### Word Sense Disambiguation

In **word sense disambiguation** you want to work out which sense of a word (or essentially which domain) was intended in the given context.

In the examples given:
- *serve*: help with food or drink; hold an office; put ball into play
- *dish*: plate; course of a meal; communications device

In the sentence *he served the dish* you can detect that both *serve* and *dish* are being used with their food meanings. This shows we automatically disambiguate words using context, exploiting the fact that nearby words have closely related meanings.

Another example of this *contextual effect*:

- The lost children were found by the *searchers* (agentive)
- The lost children were found by the *mountain* (locative)
- The lost children were found by the *afternoon* (temporal)

#### Pronoun Resolution

A deeper kind of language understanding to work out "who did *what* to *whom*",essentially detecting the subjects and objects of verbs.

Example:
- The thieves stole the paintings. They were subsequently *sold*. (The **paintings** were sold)
- The thieves stole the paintings. They were subsequently *caught*. (The **thieves** were caught)
- The thieves stole the paintings. They were subsequently *found*. (This is *ambiguous*, either the paintings or the thieves could be found)

To be able to determine the subjects and objects of verb, this involves finding the **antecedent** of the pronoun they, either thieves or paintings.

#### Brown Corpus

The Brown Corpus was the first million-word electronic corpus of English, created in 1961 at Brown University. This corpus contains text from 500 sources, and the sources have been categorized by genre, such as news, editorial, and so on.

````python
from nltk.corpus import brown
print(brown.categories())

print(brown.words(categories='news'))
print(brown.words(fileids=['cg22']))
print(brown.sents(categories=['news','editorial','reviews']))
````

````python
news_text = brown.words(categories='news')
fdist = nltk.FreqDist(w.lower() for w in news_text)

# Modal verb usage in the news category
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
    print(m + ':', fdist[m], end=' ')
````

````python
cfd = nltk.ConditionalFreqDist(
    (genre, word)
    for genre in brown.categories()
    for word in brown.words(categories=genre))

genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
````

#### Inaugural Address Corpus

One interesting property of this collection is its time dimension. We can look at the usage of words in the inaugural addresses of different presidents over time.

````python
from nltk.corpus import inaugural

# Print fileids from the corpus
print(inaugural.fileids())

# Print the years corresponding to each inaugural address
print([fileid[:4] for fileid in inaugural.fileids()])

# Plot the usage of the words America and citizen are used over time
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:4])
    for fileid in inaugural.fileids()
    for w in inaugural.words(fileid)
    for target in ['america', 'citizen']
    if w.lower().startswith(target))

cfd.plot()
````

````python
# Function for determining the plural of a word (though not always correct)

def plural(word):
    if word.endswith('y'):
        return word[:-1] + 'ies'
    elif word[-1] in 'sx' or word[-2:] in ['sh', 'ch']:
        return word + 'es'
    elif word.endswith('an'):
        return word[:-2] + 'en'
    else:
        return word + 's'
    
print(plural('farmer'))
print(plural('ruby'))
print(plural('dish'))
print(plural('woman'))

# This one will be incorrect
print(plural('fan'))
````

#### Corpus of stopwords

Stopwords are high-frequency words like the, to, and also that we sometimes want to filter out of a document before further processing. Stopwords usually have little lexical content - their presence in a text fails to distinguish it from other texts.

````python
from nltk.corpus import stopwords
print(stopwords.words('english'))
````

````python
# Count fraction of words in a text that are not in the stopwords list
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)


# Use function on Reuters corpus
print(content_fraction(nltk.corpus.reuters.words()))
````

### **WordNet**
WordNet is a semantically-oriented dictionary of English, similar to a traditional thesaurus but with a richer structure. NLTK includes the English WordNet, with 155,287 words and 117,659 synonym sets. We'll begin by looking at synonyms and how they are accessed in WordNet.

Lets take these examples:
* Benz is credited with the invention of the motorcar.
* Benz is credited with the invention of the automobile.

Since everything else in the sentence has remained unchanged we can conclude that the words *motorcar* and *automobile* havwe the same meaning, i.e. they are **synonyms**. We can explore these words with the help of WordNet.


````python
from nltk.corpus import wordnet as wn

# Find all synsets (synonym sets) of the word motorcar
wn.synsets('motorcar')
````

````python
# List the collection of synonymous words
print(wn.synset('car.n.01').lemma_names())

# Give the definition of this synset
print(wn.synset('car.n.01').definition())

# Provide an example sentence using the word car
print(wn.synset('car.n.01').examples())
````

The words of a synset are often more useful for programs. To eliminate ambiguity, we identify these words as `car.n.01.automobile`, `car.n.01.motorcar` and so on. The pairing of a synset with a word is called a **lemma**.

````python
# Get all lemmas in a synset
print(wn.synset('car.n.01').lemmas())

# Find a particular lemma
print(wn.lemma('car.n.01.automobile'))

# Get the synset corresponding to a lemma
print(wn.lemma('car.n.01.automobile').synset())

# Get the "name" of a lemma (the word)
print(wn.lemma('car.n.01.automobile').name())
````

````python
# Find all synsets of the word 'car'
print(wn.synsets('car'))

# Print all synonyms of each synset of the word 'car'
for synset in wn.synsets('car'):
    print(synset.lemma_names())

# Access all the lemmas of the word car
print(wn.lemmas('car'))
````

WordNet synsets correspond to abstract concepts,a nd they don't always have corresponding words in English. These concepts are linked together in a hierarchy. Some concepts such as *Entity, State, Event* are very general - these are called **unique beginners** or root synsets.

WordNet makes it easy to navigate between concepts. For a given concept like *motorcar* we can look at concepts that are more specific; the immediate **hyponyms**

````python
motorcar = wn.synset('car.n.01')
types_of_motorcar = motorcar.hyponyms()

print(sorted(lemma.name() for synset in types_of_motorcar for lemma in synset.lemmas()))
````

````python
# Navigafe up the hierachy by finding hypernyms
print(motorcar.hypernyms())

# Find all hypernym paths
paths = motorcar.hypernym_paths()
print(len(paths))

# Print first path
print([synset.name() for synset in paths[0]])

# Print second path
print([synset.name() for synset in paths[1]])

# Find all root hypernyms
# Print first path
print(motorcar.root_hypernyms())
````
