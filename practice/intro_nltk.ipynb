### Introduction to the Natural Language Processing Toolkit (nltk)

The following exercises follow the [NLTK book](https://www.nltk.org/book/ch01)

```python
import matplotlib
import numpy

# Import all texts from the NTLK book module
from nltk.book import *
```

```python
# Search for text using concordance view (search for occurrences)
print(f"\nSearch for monstrous in {text1}")
text1.concordance("monstrous")

print(f"\nSearch for monstrous in {text2}")
text2.concordance("monstrous")

print(f"\nSearch for Seth in {text3}")
text3.concordance("Seth")
```

```python
# Find other words that have occured in similar contexts
print(text1.similar("monstrous"))
print(text2.similar("meaning"))
```

```python
# Examine contexts that are shared by two or more words
print(text2.common_contexts(["monstrous", "very"]))
print(text3.common_contexts(["tree", "life"]))
```

```python
# Display the positional information of a word occurring in a text using a dispersion plot
text4.dispersion_plot(["citizens", "democracy", "freedom", "duties", "America"])

text3.dispersion_plot(["son", "years", "ark", "said"])
```

```python
# Generate random text in styles of the various texts
print(f"\nStyle: {text3}")
print(text3.generate())

print(f"\nStyle: {text4}")
print(text4.generate())
```

```python
# Count the length of text
print(f"Length of {text3}: {len(text3)}")
print(f"Sorted tokens {text3}: {sorted(set(text3))}")
print(f"Length of tokens {text3}: {len(set(text3))}")
```

```python
# Function to find the lexical diversity
def lexical_diversity(text):
    return len(set(text)) / len(text)

def percentage(count, total):
    return 100 * count / total
```

```python
print(f"Lexical diversity of {text1}: {lexical_diversity(text1)}")
print(f"Lexical diversity of {text2}: {lexical_diversity(text2)}")
print(f"Lexical diversity of {text3}: {lexical_diversity(text3)}")
print(f"Lexical diversity of {text4}: {lexical_diversity(text4)}")
print(f"Lexical diversity of {text5}: {lexical_diversity(text5)}")

print(f"Lexical diversity of {text5}: {lexical_diversity(text5)}")
```
